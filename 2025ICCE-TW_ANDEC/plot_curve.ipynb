```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取 Excel 檔案
file_path = "Trade-off_DEC_AN.xlsx"
xls = pd.ExcelFile(file_path)

# 讀取工作表
df = pd.read_excel(xls, sheet_name="工作表1")

# 重新命名列名
df.columns = [
    "Data bits", "Amin", "A_bits", "N_bits",
    "Parallel Area", "Parallel Time",
    "Sequential Area", "Sequential Time",
    "Trade-off Area", "Trade-off Time"
]

# 移除第一行（原來的列名）
df = df.iloc[1:].reset_index(drop=True)

# 轉換數據類型
df = df.apply(pd.to_numeric, errors='coerce')

# 設置圖表大小
plt.figure(figsize=(15, 8))

# 繪製不同的折線

plt.plot(df["Data bits"], df["Parallel Area"], marker='X', linestyle='-', color='#DC143C', label="Parallel Area", markersize=10,linewidth=3)  # red
plt.plot(df["Data bits"], df["Parallel Time"], marker='P', linestyle='-', color='#00FFFF', label="Parallel Time", markersize=10,linewidth=3, alpha=0.8)  # cyan

plt.plot(df["Data bits"], df["Sequential Area"], marker='*', linestyle='--', color='#000000', label="Sequential Area", markersize=5,linewidth=2)  # black
plt.plot(df["Data bits"], df["Sequential Time"], marker='D', linestyle='--', color='#0000FF', label="Sequential Time", markersize=4,linewidth=2, alpha=0.8)  # blue 

plt.plot(df["Data bits"], df["Trade-off Area"], marker='v', linestyle='-.', color='#FF00FF', label="Trade-off Area", markersize=4,linewidth=1)  # magenta
plt.plot(df["Data bits"], df["Trade-off Time"], marker='h', linestyle='-.', color='#FF8C00', label="Trade-off Time", markersize=5,linewidth=1, alpha=0.8)  # orange

# 添加標籤和標題
plt.xlabel("Data Bits",fontsize = 16)
plt.ylabel("Area & Time Budget",fontsize = 16)
plt.title("Parallel v.s. Sequential v.s. Trade-off", fontsize = 20)

# 設定圖例字體大小
plt.legend(fontsize=20)
plt.grid(True, linestyle="--", alpha=0.6)

# 顯示圖表
plt.show()
```
