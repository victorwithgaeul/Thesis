## 評估乘法及模除法數量:上限值,下限值,平均值

### 以鑰匙長度($p * q = n$) = 24bits , $M = (p-1)(q-1)$ ≈ 24 bits為例:


### A.產生公私鑰的過程:

#### 1. is_prime函數(輾轉相除法):
* 約有 $⌊\sqrt{number}$ - $1$⌋ 次mod運算: 此例以 p,q 為 12bits, 約為 $2^6$ 次 模除運算(位數為 $0$ ~ $6bits$ )
* 兩個數 p, q, 因此共進行 $2$ * $2^6$ = $2^7$ 次 模除運算 (位數為 $0$ ~ $6bits$ )
* In General: 2個 $\frac{n}{4}$ mod運算 = $2^{(\frac{n}{4} + 1)}$ 次 模除運算, 位數為 $0$ ~ $(\frac{n}{4}) bits$ 

#### 2.  $p * q = n$ 及 $M = (p-1)(q-1)$
* 需要 2次 12 bits x 12 bits 的乘法
* In General: 2次 $\frac{n}{2} bits * \frac{n}{2} bits$ 乘法

#### 3.公鑰取常見的值:65537, 確認 e 與 M 是否互質
* 基於歐幾里得算法, 平均情況為 $O(log_{2}(min(a,b)))$ 次 模除運算
* 在此為: $log_{2}(65537)$ ≈ 16次 模除運算

#### 4.求出私鑰 d , 並計算產生私鑰的過程中需要進行___次乘法,___次mod運算
* 結果顯示需要 (M-1)次乘法 及 (M-1) 次模除
* 最浪費時間運算
* In General: $2^{n}$ 次乘法 及 $2^{n}$ 次模除

### B. 用公鑰加密的過程: $𝑋^𝑒$  $mod$ $n$ = $C$
* 以下位元數皆以明文資料 $X$ 決定

#### 直接計算: 
* $(e-1)$ 次乘法  及   $1$ 次模除運算 
* 若公鑰為 65537, 需約 $2^{16}$ 次乘法  及   $1$ 次模除運算 

#### 快速冪法: 若 $e$ 的二進制中有 $k$ 個 1，需要額外 $(k − 1)$ 次乘法
* ≈ ⌊$log_{2}(e)$⌋ + $(k − 1)$ 次乘法 及 ⌊$log_{2}(e)$⌋ + $(k − 1)$ 次模除運算
* 每次平方或乘法後立刻對結果取模，確保數值不會超出模數 𝑛, ex: {[[(($x$ * $x$) mod $n$) * $x$] mod $n$] * $x$ ......} mod $n$
* 若公鑰為65537, 需約 ($16+1$) 次乘法  及 ($16+1$) 次模除運算

* 以下位元數皆以密文資料 $C$ 決定 (相較明文 $X$ 複雜許多)

* $(d-1)$ 次乘法  及  $1$ 次模除運算
