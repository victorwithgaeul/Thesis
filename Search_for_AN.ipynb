## Define function

```python
import math
import numpy as np
def decimal_to_binary(decimal_number):
    if decimal_number == 0:
        return "0"
    binary_number = ""
    while decimal_number > 0:
        binary_number = str(decimal_number % 2) + binary_number
        decimal_number //= 2
    return binary_number

def binary_to_decimal(binary_number):
    try:
        # 驗證輸入是否為有效的二進制數字
        decimal_number = int(binary_number, 2)
        return decimal_number
    except ValueError:
        return "輸入的不是有效的二進制數字！"


def decimal_to_binary_bits(decimal_number):
    if decimal_number == 0:
        return 1  # 0 的二進制表示為 0，只有 1 個位元
    binary_bits = 0
    while decimal_number > 0:
        binary_bits += 1
        decimal_number //= 2
    return binary_bits

def bit_range(bits):
    if bits <= 0:
        return "位元數必須為正整數！"
    min_value = 0  # 最小值
    max_value = 2**bits - 1  # 最大值
    return max_value

from math import comb
def error_combinations(binary_string):
    n = len(binary_string)  #字串長度
    r = 2  #每次取兩個數
    return comb(n, r)

def count_bits(binary_string):
    # 計算字串長度，即位元數
    return len(binary_string)

def enumerate_digits(binary_string):
    indexed_digits = [(index, char) for index, char in enumerate(binary_string)]
    return indexed_digits

def flip(x):          ## 錯誤:即翻轉bit
    return (1-x)

def binarylist_to_decimal(binary_list):
    # 將列表中的數字轉換為二進制字串
    binary_str = ''.join(str(bit) for bit in binary_list)
    # 將二進制字串轉換為十進制
    decimal_value = int(binary_str, 2)
    return decimal_value

def remainder_unique(input_list):
    # 使用集合判斷元素是否有重複
    return len(input_list) == len(set(input_list))

## 把重複的 AWE 去除
def remove_duplicates(input_list):
    unique_list = []
    for number in input_list:
        if number not in unique_list:
            unique_list.append(number)
    return sorted(unique_list)
```

## 給一固定data bits數 ,找尋可能的AN
### (此例為位元錯誤, 並非DEC_ANcode的 Double Arithmetic-Weight Error)

```python
### 輸入資料: data_bits

data_bits = 8
AN_bits = 11
data_range = bit_range(data_bits)
AN_range = bit_range(AN_bits)
print("輸入資料範圍:",data_range)
print("ANcode範圍:",AN_range)

FALSE_AN = [0] * (data_range+1)               ## 無法當ANcode, A的總數
GOOD_AN = [[] for _ in range(data_range+1)]   ## 可以成功當AN的list

for data in range(1, data_range+1):
    print("當資料為", (data-1), "時,AN的list為:\n", GOOD_AN[data-1])
    for AN in range(1, AN_range):
        input = data*AN
        binary_input = decimal_to_binary(input)
        bit_counts = count_bits(binary_input)
        combinations = error_combinations(binary_input)
        indexed_digits = enumerate_digits(binary_input)
        code=[]
        rows, cols = combinations , bit_counts                         ## 行數和列數定為 錯誤情形數 和 bits數
        error_data = [[None] * cols for _ in range(rows+bit_counts)]   ## 創一個二維錯誤資料陣列 (list)
                                                                       ## 包含2-bits error and 1-bits error
        p=0                       ## 錯誤旗標  
        decimal_error_data = []   ## 所有可能2bits error的錯誤資料(十進制)
        remainder = []            ## 所有可能的餘數
        for index, char in indexed_digits:
            code.append(int(char))
        ### 產生2-bits error資料:
        ####################################################################################
        for i in range(0, index+1):
            for j in range(i+1, index+1):
                error_data[p] = [code[k] for k in range(len(code)) if (k != i) & (k != j)]  # 刪掉2個錯誤的bits,位置分別在 i,j
                error_data[p].insert(i,flip(code[i]))                                       # 插入錯的 bit-1 (位置在i)
                error_data[p].insert(j,flip(code[j]))                                       # 插入錯的 bit-2 (位置在j)
                ### print(f"2-bits第 {p} 筆錯誤資料: {error_data[p]}")
                p=p+1
        ### 產生1-bits error資料:
        for i in range(0, index+1):
                error_data[p] = [code[k] for k in range(len(code)) if (k != i)]      ## 刪掉1個錯誤的bit, 位置分別在 i
                error_data[p].insert(i,flip(code[i]))                                ## 插入錯的 bit-1 (位置在i)
                ###　print(f"1-bits第 {p} 筆錯誤資料: {error_data[p]}")
                p=p+1
            
        ## 將error_dat轉為十進制
        ####################################################################################
        for i in range(0,p):
            decimal_error_data.append(binarylist_to_decimal(error_data[i]))
        ## 把重複的 AWE 去除
        decimal_error_data = remove_duplicates(decimal_error_data)
        ## 產生所有error_data模AN的餘數
        ####################################################################################
        for m in range(0,p):
            mod = decimal_error_data[m] % AN
            remainder.append(mod)
        ## 確認所有的餘數是否完全相異
        ####################################################################################
        if remainder_unique(remainder):
            ###　print("當輸入=", data,"AN=", AN,"時,每筆錯誤資料的模數皆不相同" )
            GOOD_AN[data].append(AN)
        else:
            FALSE_AN[data] = FALSE_AN[data] + 1
        ####################################################################################


print("原始資料:", code)
del GOOD_AN[0]         ### 刪掉多的空集合
###  print("無法當AN的A數量為:", FALSE_AN)
```

```python
import math
import numpy as np

def decimal_to_binary(decimal_number):
    if decimal_number == 0:
        return "0"
    # 處理正數：直接轉換
    if decimal_number > 0:
        return bin(decimal_number)[2:]  # 移除 "0b" 前綴
       # 處理負數：自動計算所需位元數（2's complement）
    abs_value = abs(decimal_number)
    bits = abs_value.bit_length() + 1  # 負數多一個符號位元
    two_complement_value = (1 << bits) + decimal_number
    
    return bin(two_complement_value)[2:]  # 移除 "0b" 前綴

def bits_needed(number):
    if number < 0:
        raise ValueError("2's complement is not available")  # 如果輸入負數，拋出錯誤
    if number == 0:
        return 1  # 0 需要 1 個位元
    
    # 計算位元數
    bits = math.floor(math.log2(number)) + 1
    return bits


def bits_range(bits, signed):     ## 可選擇設定 有號數 或 無號數
    if bits <= 0:
        return "位元數必須為正整數！"
    
    if signed:
        # 二補數範圍: -2^(bits-1) 到 2^(bits-1)-1
        min_value = -(2**(bits - 1))
        max_value = 2**(bits - 1) - 1
    else:
        # 無符號範圍: 0 到 2^bits - 1
        min_value = 0
        max_value = 2**bits - 1
    return min_value, max_value

## 把重複的 AWE 去除
def remove_duplicates(input_list):
    unique_list = []
    for number in input_list:
        if number not in unique_list:
            unique_list.append(number)
    return sorted(unique_list)

def custom_mod(a, b):
    if b == 0:
        return "錯誤：除數不能為零！"
    # 計算商和餘數，修正餘數計算邏輯
    quotient = a // b
    remainder = a - b * quotient
    # 修正負餘數的情況
    if remainder > 0 and a < 0:
        remainder -= abs(b)
        quotient += 1
    return remainder

def remainder_unique(input_list):
    # 使用集合判斷元素是否有重複
    return len(input_list) == len(set(input_list))

print(custom_mod(-90,8))
```

```python
import math
import numpy as np

def decimal_to_binary(decimal_number):
    if decimal_number == 0:
        return "0"
    # 處理正數：直接轉換
    if decimal_number > 0:
        return bin(decimal_number)[2:]  # 移除 "0b" 前綴
       # 處理負數：自動計算所需位元數（2's complement）
    abs_value = abs(decimal_number)
    bits = abs_value.bit_length() + 1  # 負數多一個符號位元
    two_complement_value = (1 << bits) + decimal_number
    
    return bin(two_complement_value)[2:]  # 移除 "0b" 前綴

def bits_needed(number):
    if number == 0:
        return 1  # 0 需要 1 個位元
    # 計算絕對值的位元數
    abs_value = abs(number)
    bits = math.floor(math.log2(abs_value)) + 1
    if number > 0:
        # 正數需要多一個符號位，避免 MSB 與符號衝突
        if 2 ** (bits - 1) <= number:
            bits += 1
    else:
        # 負數需要符號位，自然補充
        bits += 1
    return bits

def bits_range(bits, signed):     ## 可選擇設定 有號數 或 無號數
    if bits <= 0:
        return "位元數必須為正整數！"
    
    if signed:
        # 二補數範圍: -2^(bits-1) 到 2^(bits-1)-1
        min_value = -(2**(bits - 1))
        max_value = 2**(bits - 1) - 1
    else:
        # 無符號範圍: 0 到 2^bits - 1
        min_value = 0
        max_value = 2**bits - 1
    return min_value, max_value

## 把重複的 AWE 去除
def remove_duplicates(input_list):
    unique_list = []
    for number in input_list:
        if number not in unique_list:
            unique_list.append(number)
    return sorted(unique_list)

def custom_mod(a, b):
    if b == 0:
        return "錯誤：除數不能為零！"
    # 計算商和餘數，修正餘數計算邏輯
    quotient = a // b
    remainder = a - b * quotient
    # 修正負餘數的情況
    if remainder > 0 and a < 0:
        remainder -= abs(b)
        quotient += 1
    return remainder

def remainder_unique(input_list):
    # 使用集合判斷元素是否有重複
    return len(input_list) == len(set(input_list))
```
