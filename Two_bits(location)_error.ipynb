## Find prime number

```python
import math
def is_prime(n):
    if n < 2:
        return False  ### 排除小於 2 的數
    i = 2
    while (i * i <= n):
        if n % i == 0:  ### 如果 n 可以被 i 整除，則 n 不是質數
            return False
        i += 1
    return True

def primes(N):
    p = []
    for i in range(2, N+1):
        if is_prime(i):
            p.append(i)
    return p

print(primes(319))
```

#### Convert between decimal and binary

```python
def decimal_to_binary(decimal_number):
    if decimal_number == 0:
        return "0"
    binary_number = ""
    ### is_negative = decimal_number < 0
    ### if is_negative:
    ### decimal_number = abs(decimal_number)
    ### if is_negative:
    ### binary_number = "-" + binary_number
    while decimal_number > 0:
        binary_number = str(decimal_number % 2) + binary_number
        decimal_number //= 2
    return binary_number

def binary_to_decimal(binary_number):
    try:
        # 驗證輸入是否為有效的二進制數字
        decimal_number = int(binary_number, 2)
        return decimal_number
    except ValueError:
        return "輸入的不是有效的二進制數字！"


def decimal_to_binary_bits(decimal_number):
    if decimal_number == 0:
        return 1  # 0 的二進制表示為 0，只有 1 個位元
    binary_bits = 0
    while decimal_number > 0:
        binary_bits += 1
        decimal_number //= 2
    return binary_bits

def bit_range(bits):
    if bits <= 0:
        return "位元數必須為正整數！"
    min_value = 0  # 最小值
    max_value = 2**bits - 1  # 最大值
    return min_value, max_value

def count_bits(binary_string):
    # 計算字串長度，即位元數
    return len(binary_string)


print(decimal_to_binary(500))
print(decimal_to_binary_bits(500))
print(binary_to_decimal(decimal_to_binary(500)))
bit = 11
min_value, max_value = bit_range(bit)
print(f"{bit} 位元可以表示的十進制範圍是: {min_value} 到 {max_value}")
```

### 在 n 位元中發生2個錯誤的情形

```python
### 輸入數值為data, ANcode的A為AN
### 乘完A後的數值為 input
data = 500
AN = 319
input = data*AN
binary_input = decimal_to_binary(input)
bit_counts = count_bits(binary_input)

from math import comb
def error_combinations(binary_string):
    n = len(binary_string)  #字串長度
    r = 2  #每次取兩個數
    return comb(n, r)

def enumerate_digits(binary_string):
    indexed_digits = [(index, char) for index, char in enumerate(binary_string)]
    return indexed_digits

combinations = error_combinations(binary_input)
indexed_digits = enumerate_digits(binary_input)
print(f"資料 {binary_input} 發生2個錯誤，有 {combinations} 種情形")
print(f"資料 {binary_input} 發生1個錯誤，有 {bit_counts} 種情形")
print(f"資料發生錯誤的所有情形，有 {(combinations+bit_counts)} 種情形")

print("\n資料(code)數字與編號對應如下：")
code=[]
for index, char in indexed_digits:
    print(f"第 {index} 個位置: {char}")
    code.append(int(char))

print(code)
print("位元數為:",index+1)
print(type(code))
```

### Error data in decimal

```python
def binarylist_to_decimal(binary_list):
    # 將列表中的數字轉換為二進制字串
    binary_str = ''.join(str(bit) for bit in binary_list)
    # 將二進制字串轉換為十進制
    decimal_value = int(binary_str, 2)
    return decimal_value

decimal_error_data = []
for i in range(0,p):
    decimal_error_data.append(binarylist_to_decimal(error_data[i]))


print("錯誤資料的十進制為:\n",decimal_error_data)

remainder = []
for m in range(0,p):
    mod = decimal_error_data[m] % AN
    remainder.append(mod)

print("錯誤資料對AN取完模數:\n",remainder)

def remainder_unique(input_list):
    # 使用集合判斷元素是否有重複
    return len(input_list) == len(set(input_list))


print("所有餘數是否完全相異:", remainder_unique(remainder))
print(np.size(remainder))
```
